#### Crawling Data From PTA Trunojoyo University
this code have a task to crawling all journall in pta.trunojoyo.ac.id that have 41 study program. hasil dari program ini disimpan dalam bentuk 1 csv yang dalam data tersebut berisi kurang lebih 14000 journal

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def ptaa():
  # Membuat dictionary untuk menyimpan data
    data = {"nama_prodi": [],"penulis": [], "judul": [], "pembimbing_pertama": [], "pembimbing_kedua": [], "abstrak": [], "abstrak_en": []}
            # Loop berdasarkan program studi (1-40, misalnya)
    for i in range(28,41):
        url = f"https://pta.trunojoyo.ac.id/c_search/byprod/{i}"
        r = requests.get(url)
        soup = BeautifulSoup(r.content, "html.parser")

        nama_prodi_tag = soup.select_one("div#begin h2")
        nama_prodi = nama_prodi_tag.text.strip() if nama_prodi_tag else f"Prodi {i}"
        print(f"Kategori {i} → {nama_prodi}")

        # cari jumlah halaman (pagination)
        page_links = soup.select("a.pag_button")
        if page_links:
            jumlahPage = int(page_links[-1]["href"].split("/")[-1])  # ambil angka terakhir
        else:
            jumlahPage = 1  # kalau cuma 1 halaman
        print(f"Kategori {i} memiliki {jumlahPage} halaman.")

        # loop tiap halaman
        for j in range(1, jumlahPage + 1):
            print("literasi ke",j)
            url = f"https://pta.trunojoyo.ac.id/c_search/byprod/{i}/{j}"
            r = requests.get(url)
            request = r.content
            soup = BeautifulSoup(request, "html.parser")
            jurnals = soup.select('li[data-cat="#luxury"]')
            print(f"Jumlah jurnal pada halaman {j}: {len(jurnals)}")

            for jurnal in jurnals:
                response = requests.get(jurnal.select_one('a.gray.button')['href']) #loncat ke halaman detail
                soup1 = BeautifulSoup(response.content, "html.parser")  #parsing halaman detail
                isi = soup1.select_one('div#content_journal')
                print("berjalan")

                isi = soup1.select_one("div#content_journal")

                # Judul
                judul = isi.select_one("a.title").text.strip() if isi.select_one("a.title") else ""


                # Penulis
                penulis = ""
                for span in isi.find_all("span"):
                    if "Penulis" in span.text:
                        penulis = span.text.split(" : ")[-1].strip()
                print(nama_prodi)
                # Pembimbing 1
                pembimbing_pertama = ""
                for span in isi.find_all("span"):
                    if "Dosen Pembimbing I" in span.text:
                        pembimbing_pertama = span.text.split(" : ")[-1].strip()

                # Pembimbing 2
                pembimbing_kedua = ""
                for span in isi.find_all("span"):
                    if "Dosen Pembimbing II" in span.text:
                        pembimbing_kedua = span.text.split(" : ")[-1].strip()

                # Abstrak
                abstrak = ""
                abstrak_en = ""
                paras = isi.find_all("p", align="justify")
                if len(paras) >= 1:
                    abstrak = paras[0].get_text(strip=True)
                if len(paras) >= 2:
                    abstrak_en = paras[1].get_text(strip=True)

                data["nama_prodi"].append(nama_prodi)
                data["penulis"].append(penulis)
                data["judul"].append(judul)
                data["pembimbing_pertama"].append(pembimbing_pertama)
                data["pembimbing_kedua"].append(pembimbing_kedua)
                data["abstrak"].append(abstrak)
                data["abstrak_en"].append(abstrak_en)
                print(len(data["judul"]))

        dfSementara = pd.DataFrame(data)
        dfSementara.to_csv(f"pta3_prodi_{i}.csv", index=False, encoding="utf-8-sig")
        print(f"Sementara disimpan untuk Prodi {i}")


    df = pd.DataFrame(data)
    df.to_csv("pta3.csv", index=False, encoding="utf-8-sig")
    return df


# Jalankan
df_hasil = ptaa()
print(df_hasil.head())
```
